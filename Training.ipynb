```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.schema import Document

def load_text_file(txt_file):
    """Reads a .txt file and returns it as a list of Document objects"""
    with open(txt_file, "r", encoding="utf-8") as file:
        text = file.read()

    return [Document(page_content=text)]  # Convert to Document format

def text_split(extracted_data):
    """Splits the document into chunks"""
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=540, chunk_overlap=20)
    text_chunks = text_splitter.split_documents(extracted_data)
    return text_chunks

# Usage Example
extracted_data = load_text_file("data.txt")
chunks = text_split(extracted_data)
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
def download_hugging_face_embeddings():
    embeddings=HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
    return embeddings
```
